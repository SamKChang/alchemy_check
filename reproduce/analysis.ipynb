```python
import qctoolkit as qtk
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
[r, t, p], R, [re, te, pe], [res, tes, pes] = qtk.load('production_shifted_results.pkl')
```

```python
R = np.array([o.molecule.R[1,0] - o.molecule.R[0,0] for o in r])
```

```python
plt.plot(R, res, label='HF')
plt.plot(R, tes, label='HCl_tar')
plt.plot(R, pes, label='HCl_prd')

plt.legend()
plt.ylabel(r'$\Delta E$ [Ha]')
plt.xlabel(r'$d$ [$\mathrm{\AA}$]')
```

```python
plt.plot(R, re, label='HF')
plt.plot(R, te, label='HCl_tar')
plt.plot(R, pe, label='HCl_prd')

plt.legend()
plt.ylabel(r'$E_{tot}$ [Ha]')
plt.xlabel(r'$d$ [$\mathrm{\AA}$]')
```
